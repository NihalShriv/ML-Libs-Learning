```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/natural-disaster-data/number-of-natural-disaster-events.csv')
```

```python
df.head()
```

```python
df['Entity'].value_counts()
```

```python
df.drop(['Code'], axis=1, inplace=True)
```

```python
df.rename(columns={'Number of reported natural disasters (reported disasters)': 'Reported Disasters'}, inplace=True)
```

```python
df.Year.value_counts(ascending=True)
```

```python
df.sort_values(by=['Year'], ascending=True, inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('index', axis=1, inplace=True)
```
