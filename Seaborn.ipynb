```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('https://raw.githubusercontent.com/NihalShriv/ML-Libs-Learning/master/Dataset/Pokemon.csv?token=AHUGWIAJIXXMC3PU6ACD7YK7G6RMA', encoding = 'unicode_escape')
```

```
df.sort_values(by='Defense')
```

```
# scatter plot
sns.scatterplot(x = 'Attack', y = 'Defense', data=df, hue='Stage', palette="bright")
plt.xlim(0)
plt.ylim(0)
plt.show()
```

```
fig, ax = plt.subplots(1,2, figsize=(20,6))

# box plot
sns.boxplot(data=df.drop(['Total', 'Stage', 'Legendary'], axis=1), ax=ax[0], palette='bright')
ax[0].set_ylim(0)

# violin plot
sns.violinplot(data=df.drop(['Total', 'Stage', 'Legendary'], axis=1), ax=ax[1], palette='bright')
ax[1].set_ylim(0)
plt.show()
```

```
fig, ax = plt.subplots(2,1, figsize=(15,10))
sns.violinplot(x='Type 1', y='Attack', data=df, palette='colorblind', ax=ax[0])
sns.violinplot(x='Type 1', y='Defense', data=df, palette='colorblind', ax=ax[1])
plt.show()
```

```
# swarm plot
# it is same as box or violin plot. the only difference is that they display whole data together whereas swarmplot displays each data point

fig, ax = plt.subplots(2,1, figsize=(15,10))
sns.swarmplot(x='Type 1', y='Attack', data=df, palette='bright', ax=ax[0])
ax[0].set_ylim(0)
ax[0].set_xlabel(xlabel='Type 1', color='grey')
ax[0].set_ylabel(ylabel='Attack', color='grey')

sns.swarmplot(x='Type 1', y='Defense', data=df, palette='bright', ax=ax[1])
ax[1].set_ylim(0)
ax[1].set_xlabel(xlabel='Type 1', color='grey')
ax[1].set_ylabel(ylabel='Defense', color='grey')
plt.show()
```

```
# overlaying plots
plt.figure(figsize=(20,6))

sns.violinplot(x='Type 1', y='Attack', data=df, palette='bright')
sns.swarmplot(x='Type 1', y='Attack', data=df, color='white')
plt.show()
```

```
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat",
                    value_name="Value") # Name of melted variable
melted_df.head()
```

```
plt.figure(figsize=(15,5))
sns.swarmplot(x='Stat', y='Value', data=melted_df, hue='Type 1', palette='bright', dodge=True)
plt.ylim(0,260)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```
sns.heatmap(stats_df.corr(), linewidths=.5)
plt.show()
```

```
plt.figure(figsize=(15,5))
sns.distplot(df.Attack, color='blue')
plt.xlim(0)
plt.show()
```
